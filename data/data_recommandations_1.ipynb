```python
import pandas as pd
import numpy as np
```

```python
#Importer les fichiers csv comme dataframes 
df_tmdb = pd.read_csv('tmdb_full.csv', sep = ',', chunksize= 10000000)
df_principals = pd.read_csv('https://datasets.imdbws.com/title.principals.tsv.gz', sep = "\t", chunksize= 10000000)
df_title_akas = pd.read_csv('https://datasets.imdbws.com/title.akas.tsv.gz', sep = "\t", chunksize= 10000000)
df_name_basics = pd.read_csv('https://datasets.imdbws.com/name.basics.tsv.gz', sep = "\t", chunksize= 10000000)
df_crew = pd.read_csv('https://datasets.imdbws.com/title.crew.tsv.gz', sep = "\t", chunksize= 10000000)
df_title_basics = pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz', sep = "\t", chunksize= 10000000)
df_ratings = pd.read_csv('https://datasets.imdbws.com/title.ratings.tsv.gz', sep = "\t", chunksize= 10000000)
```

```python
#Créer Dataframe tmdb fr
#Liste vide à compléter
df_tmdb_fr = []

#Parcourir chaque chunk et nettoyer les données
for chunk in df_tmdb:
    #Filtrer les lignes où la colonne 'original_language' est 'fr'
    chunk_tmdb_clean = chunk.loc[chunk['original_language'] == 'fr']
    df_tmdb_fr.append(chunk_tmdb_clean)

#Concaténer les chunks nettoyés en un seul DataFrame
df_tmdb_fr_1 = pd.concat(df_tmdb_fr)

#Afficher le DataFrame résultant
pd.options.display.max_columns = None
df_tmdb_fr_1.head(1)
```

```python
df_tmdb_fr_2 = df_tmdb_fr_1[['imdb_id', 'original_language', 'overview', 'poster_path', 'original_title', 'vote_average', 'vote_count']]
df_tmdb_fr_2.head()
```

```python
#Créer Dataframe titres fr
#Liste vide à compléter
df_titres_fr = []

#Parcourir chaque chunk et nettoyer les données
for chunk in df_title_basics:
    #Filtrer les lignes où la colonne 'tconst' est == 'imdb_id'
    chunk_titres_clean = chunk.loc[chunk["tconst"].isin(df_tmdb_fr_2["imdb_id"])].copy()
    chunk_withstartyear = chunk_titres_clean.loc[chunk_titres_clean["startYear"] != "\\N"].copy()  # drop all titles that don't have a year listed
    filtered_chunk = chunk_withstartyear.loc[(chunk_withstartyear["titleType"] == "movie")
                                      & (chunk_withstartyear["startYear"].astype(int) >= 1970)]
    df_titres_fr.append(filtered_chunk)

#Concaténer les chunks nettoyés en un seul DataFrame
df_titres_fr_1 = pd.concat(df_titres_fr)

#Afficher le DataFrame résultant
pd.options.display.max_columns = None
df_titres_fr_1.head(1)
```

```python
#df_titres
df_titres_fr_2 = df_titres_fr_1[['tconst', 'startYear', 'genres']]
df_titres_fr_2.head()
```

```python
#Créer Dataframe acteurs fr
#Liste vide à compléter
df_acteurs_fr = []

#Parcourir chaque chunk et nettoyer les données
for chunk in df_principals:
    #Filtrer les lignes où la colonne 'tconst' est == 'imdb_id'
    chunk_acteurs_clean = chunk.loc[chunk["tconst"].isin(df_titres_fr_2["tconst"])]
    df_acteurs_fr.append(chunk_acteurs_clean)

#Concaténer les chunks nettoyés en un seul DataFrame
df_acteurs_fr_1 = pd.concat(df_acteurs_fr)

#Afficher le DataFrame résultant
pd.options.display.max_columns = None
df_acteurs_fr_1.head(1)
```

```python
#Créer Dataframe réalisateurs fr
#Liste vide à compléter
df_realisateurs_fr = []

#Parcourir chaque chunk et nettoyer les données
for chunk in df_crew:
    #Filtrer les lignes où la colonne 'tconst' est == 'imdb_id'
    chunk_realisateurs_clean = chunk.loc[chunk["tconst"].isin(df_titres_fr_2["tconst"])]
    df_realisateurs_fr.append(chunk_realisateurs_clean)

#Concaténer les chunks nettoyés en un seul DataFrame
df_realisateurs_fr_1 = pd.concat(df_realisateurs_fr)

#Afficher le DataFrame résultant
pd.options.display.max_columns = None
df_realisateurs_fr_1.head(1)
```

```python
#Créer Dataframe equipes fr
#Liste vide à compléter
df_equipes_fr = []

#Parcourir chaque chunk et nettoyer les données
for chunk in df_name_basics:
    #Filtrer les lignes où la colonne 'tconst' est == 'imdb_id'
    chunk_equipes_clean = chunk.loc[chunk["nconst"].isin(df_acteurs_fr_1["nconst"])]
    df_equipes_fr.append(chunk_equipes_clean)

#Concaténer les chunks nettoyés en un seul DataFrame
df_equipes_fr_1 = pd.concat(df_equipes_fr)

#Afficher le DataFrame résultant
pd.options.display.max_columns = None
df_equipes_fr_1.head(1)
```

```python
#df_equipes_fr_2
df_equipes_fr_2 = df_equipes_fr_1[['nconst', 'primaryName']]
df_equipes_fr_2.head()
```

```python
#df_realisateurs_fr_2
df_realisateurs_fr_2 = pd.merge(df_acteurs_fr_1[df_acteurs_fr_1['category'] == 'director'], df_equipes_fr_2, on='nconst')
```

```python
#df_acteurs_fr_2
df_acteurs_fr_2 = pd.merge(df_acteurs_fr_1[df_acteurs_fr_1['category'] == 'actor'], df_equipes_fr_2, on='nconst')
df_acteurs_fr_2.head()
```

```python
#df_acteurs_fr_3
df_acteurs_fr_3 = pd.merge(df_acteurs_fr_1[df_acteurs_fr_1['category'] == 'actress'], df_equipes_fr_2, on='nconst')
df_acteurs_fr_3.head()
```

```python
#Fusionner les différents dataframes 
df_fr_merged_1 = pd.merge(df_titres_fr_2, df_acteurs_fr_2, on='tconst')
df_fr_merged_1.head()
```

```python
df_fr_merged_2 = pd.merge(df_fr_merged_1, df_acteurs_fr_3, on='tconst')
df_fr_merged_2.head()
```

```python
df_fr_merged_3 = pd.merge(df_fr_merged_2, df_realisateurs_fr_2, on='tconst')
df_fr_merged_3.head()
```

```python
df_fr_merged_4 = pd.merge(df_fr_merged_3, df_tmdb_fr_2, how="left", left_on='tconst', right_on='imdb_id')
df_fr_merged_4.head()
```

```python
#Supprimer les colonnes en doublon ou non utiles
df_fr_merged_5 = df_fr_merged_4.drop(['ordering', 'nconst', 'category', 'job', 'characters', 
                                      'ordering_x', 'nconst_x', 'category_x', 'job_x', 'characters_x',
                                      'ordering_y', 'nconst_y', 'category_y', 'job_y', 'characters_y', 
                                      'imdb_id'], axis=1)
df_fr_merged_5.head()
```

```python
#Grouper les acteurs et realisateurs par film
df_fr_merged_6 = df_fr_merged_5.groupby(['tconst', 
                                         'startYear', 
                                         'genres', 
                                         'original_title', 
                                         'vote_average', 
                                         'vote_count', 
                                         'original_language', 
                                         'overview', 
                                         'poster_path'])[['primaryName', 'primaryName_x', 'primaryName_y']].agg(lambda x: ', '.join(x.astype(str))).reset_index()
df_fr_merged_6.head()
```

```python
#Enlever les doublons dans les colonnes
# Fonction pour supprimer les doublons d'une liste tout en préservant l'ordre
def remove_duplicates_keep_order(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]

# Appliquer la fonction à chaque cellule des colonnes 'primaryName_x' et 'primaryName_y'
df_fr_merged_6['primaryName'] = df_fr_merged_6['primaryName'].apply(lambda x: ', '.join(remove_duplicates_keep_order(x.split(', '))))
df_fr_merged_6['primaryName_x'] = df_fr_merged_6['primaryName_x'].apply(lambda x: ', '.join(remove_duplicates_keep_order(x.split(', '))))
df_fr_merged_6['primaryName_y'] = df_fr_merged_6['primaryName_y'].apply(lambda x: ', '.join(remove_duplicates_keep_order(x.split(', '))))
```

```python
df_fr_merged_6['primaryName_x'] = df_fr_merged_6['primaryName_x'] + df_fr_merged_6['primaryName_y']
df_fr_merged_7 = df_fr_merged_6.drop(columns='primaryName_y', axis= 1)
```

```python
#Créer dataframe films fr
#Organiser les colonnes
df_fr_merged_8 = df_fr_merged_7[['tconst', 'original_title', 'startYear', 'original_language', 'primaryName', 'primaryName_x', 'vote_average', 'vote_count', 'genres', 'overview', 'poster_path']]
df_fr_merged_8.head()
```

```python
#Renommer les colonnes
df_films_fr = df_fr_merged_8.rename(columns={'tconst': 'id', 
                                             'original_title': 'titre',
                                             'startYear' : 'annee',
                                             'original_language' : 'langue_originale',
                                             'primaryName' : 'realisateurs',
                                             'primaryName_x' : 'acteurs',
                                             'vote_average' : 'note',
                                             'vote_count' : 'nombre_votants',
                                             'genres' : 'genres', 
                                             'overview' : 'synopsis',
                                             'poster_path' : 'affiche'})
df_films_fr.head()
```

```python
#Exporter dataframe films fr en csv
df_films_fr.to_csv("data_recommandations_fr.csv", sep = "\t", encoding = "utf-8", index = False)
```
