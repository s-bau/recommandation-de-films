```python
import pandas as pd
import numpy as np
```

```python
#Importer les fichiers csv comme dataframes 
df_principals = pd.read_csv('data/title_principals_clean.csv', sep='\t')
df_title_akas = pd.read_csv('data/title_akas_clean.csv', sep='\t')
df_name_basics = pd.read_csv('data/name_basics_clean.csv', sep='\t')
df_crew = pd.read_csv('data/title_crew_clean.csv', sep='\t')
df_title_basics = pd.read_csv('data/title_basics_clean.csv', sep='\t')
df_ratings = pd.read_csv('data/title_ratings_clean.csv', sep='\t')
df_tmdb = pd.read_csv('data/tmdb_clean.csv', sep='\t')
```

```python
#df_titres
df_titres = df_title_basics.copy()
df_titres.head()
```

```python
#df_infos
df_infos = df_tmdb.copy()
df_infos.head()
```

```python
#Créer dataframe genres
#Conserver uniquement les colonnes souhaitées dans le dataframe
df_titres_1 = df_titres[['tconst', 'originalTitle', 'startYear', 'genres']]

#Séparer les différents genres au sein de la colonne 'genres' 
df_titres_1['genres'] = df_titres_1['genres'].str.split(',')

#Utiliser la fonction explode pour obtenir une ligne par genre
#Utiliser la fonction reset_index pour dupliquer les autres colonnes
df_titres_1 = df_titres_1.explode('genres').reset_index(drop=True)

#Utiliser la fonction head pour avoir un aperçu des 10 premières lignes
df_titres_1.head(10)
```

```python
#Définir classement des genres pour le comparer avec résultats dans Tableau 
#Utiliser la fonction value_counts pour compter le nombre de répétition d'un genre 
genre_counts = df_titres_1['genres'].value_counts()
print(genre_counts)
```

```python
#Exporter df_titres_1 en csv comme df_genres
#Travailler sur le classement dans Tableau 
df_genres = pd.DataFrame(df_titres_1)
df_genres.to_csv("dashboard_genres.csv", sep = ",", encoding = "utf-8", index = False)
df_genres
```

```python
#Créer dataframe rentabilité 
#Sélectionner les différents dataframes
df_renta = pd.merge(df_infos, df_titres, how="left", left_on='imdb_id', right_on='tconst')
#df_renta.head()

#Sélectionner les colonnes souhaitées dans les différents dataframes
df_renta_1 = df_renta[['tconst', 'originalTitle', 'startYear', 'vote_average', 'budget', 'revenue']]

#Créer colonne rentabilité
df_renta_1["rentabilite"] = df_renta_1["revenue"] - df_renta_1["budget"]
#df_renta_1.head()
```

```python
#Définir classement des films par recettes pour le comparer avec résultats dans Tableau 
df_renta_1.sort_values('rentabilite', ascending= False)
```

```python
#Définir classement des films par rentabilité pour le comparer avec résultats dans Tableau 
df_renta_1.sort_values('revenue', ascending= False)
```

```python
# checking where startYear is NaN
print(df_renta[df_renta["startYear"].isna()])
```

```python
# dropping 11 rows of NaN startyear
df_renta.dropna(subset=["startYear"], inplace=True)

df_renta["startYear"] = df_renta["startYear"].apply(lambda x: int(x))
print(df_renta.head())
```

```python
#Sélectionner les colonnes souhaitées dans les différents dataframes
df_renta_1 = df_renta[['tconst', 'originalTitle', 'startYear', 'vote_average', 'budget', 'revenue']]
print(df_renta_1.head())
```

```python
#Créer colonne rentabilité
df_renta_1["rentabilite"] = df_renta_1["revenue"] - df_renta_1["budget"]
print(df_renta_1.head())
```

```python
# exporter en csv
df_rentabilite = pd.DataFrame(df_renta_1)
df_rentabilite.to_csv("dashboard_revenue.csv", sep = ",", encoding = "utf-8", index = False)
```
